```python
import numpy as np
import pandas as pd
import bqplot as bq
import ipywidgets as widgets

def f(x):
    return x**2

# Reemplaza esta función con tu implementación real
def SumaSupSumaInf(f, a, b, paso):
    xs = np.arange(a, b+paso, paso)
    sup = [sum(f(x + paso) for x in xs[:-1])]
    inf = [sum(f(x) for x in xs[:-1])]
    integral = [sum(f(x + paso/2) for x in xs[:-1])]
    return {
        "Rectángulos": list(range(1, len(xs))),
        "Suma Inferior": inf * (len(xs)-1),
        "Suma Superior": sup * (len(xs)-1),
        "Integral": integral * (len(xs)-1)
    }

datitos = SumaSupSumaInf(f, 0, 2, 0.1)
datos = pd.DataFrame(datitos)

val_x = datos["Rectángulos"]
val_y_inf = datos["Suma Inferior"]
val_y_sup = datos["Suma Superior"]
val_y_int = datos["Integral"]

x_sc = bq.LinearScale()
y_sc = bq.LinearScale()

line_inf = bq.Lines(
    x=val_x, y=val_y_inf,
    scales={'x': x_sc, 'y': y_sc},
    colors=["blue"], stroke_width=2,
    markers=True, marker_size=64,
    labels=["Suma inferior"]
)

line_int = bq.Lines(
    x=val_x, y=val_y_int,
    scales={'x': x_sc, 'y': y_sc},
    colors=["red"], stroke_width=2,
    markers=True, marker_size=64,
    labels=["Integral"]
)

line_sup = bq.Lines(
    x=val_x, y=val_y_sup,
    scales={'x': x_sc, 'y': y_sc},
    colors=["green"], stroke_width=2,
    markers=True, marker_size=64,
    labels=["Suma superior"]
)

ax_x = bq.Axis(label='Número de rectángulos (N)', scale=x_sc)
ax_y = bq.Axis(label='Suma', scale=y_sc, orientation='vertical')

fig = bq.Figure(
    marks=[line_inf, line_int, line_sup],
    axes=[ax_x, ax_y],
    title="Convergencia de las Sumas de Riemann",
    legend_location='top-left'
)

fig
```
